```python
from sklearn.preprocessing import RobustScaler
scale = RobustScaler() 
traindropped = pd.DataFrame(traindropped).fillna(value = 0)
traindropped = scale.fit_transform(traindropped)
# np.where(np.isnan(traindropped))
```

```python
from sklearn.decomposition import PCA

pca = PCA().fit(traindropped) # whiten=True
feature_len = 3

print('With only {} features: {:6.4%} variance'.format(feature_len, sum(pca.explained_variance_ratio_[:feature_len])),"\n")

pca = PCA(sum(pca.explained_variance_ratio_[:feature_len]))
principal_components = pca.fit_transform(traindropped)
principal_df = pd.DataFrame(data = principal_components)
principal_df
```
